#Univariate Analysis



Univariate analysis involves examining a single variable in isolation. It helps in understanding the distribution, central tendency, and variability of the variable. Common techniques include:

#Histogram & Boxplot

```python
plt.figure(figsize=(8,6))
plt.subplot(2,2,1)
x1=data['Votes'].tail(100)
plt.hist(x1,bins=8,color='skyblue',
edgecolor='k')
plt.title("votes over 100 records")


plt.subplot(2,2,2)
data['Votes'].head(100).plot.box()
plt.show()
```

Conclusion 1:
- In 1st graph only last 100 records are considered.
- In histogram we can clearly see that there are many votes in 0 to 200 range.
- Over 9551 restaurants only 1 restaurant got vote above 1000, which is an outlier.
- Same as previous graph median showing range between 0 to 200.

#Histogram

```python
data['Aggregate rating'].head(20).plot.hist(bins=6,edgecolor='k',color='g')
plt.title('Frequency of Aggregate rating')
plt.show()
```

Conclusion 2:
